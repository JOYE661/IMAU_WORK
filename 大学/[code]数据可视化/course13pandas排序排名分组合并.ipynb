#### 　　3.2.2.2 排名

```python
'''
排名rank()方法是指从1开始一直到数据中的有效数据的数量为止，返回数据在排序中的位置。
'''
rrank=pd.Series([10,12,9,9,14,4,2,4,9,1])
rrank
```

```python
rrank.rank()#默认升序，相同排名取平均值
```

```python
rrank.rank(ascending=False)
```

```python
'''
当有多个数据值相同时，会出现排名也相同的情况，这时可使用rank方法的method参数来处理：
-- average:默认，在相同排名中为各个值平均分配排名
-- min：使用整个相同排名的最小排名
-- max：使用整个相同排名的最大排名
-- first：按值在原始数据中的出现顺序分配排名
'''
```

```python
rrank.rank(method='first')
```

```python
rrank.rank(method='max')
```

```python
scoresheet2.rank() #返回在各自列中排序的位置
```

#### 　　3.2.2.3 运算

```python
'''
pandas最重要的一个功能便是对不同索引的对象进行算术运算，而且在运算时系统会按照各自索引
进行自动对齐。
如果在参加运算的数据中存在不同的索引时，则结果的索引就是所有索引的并集，对应不同索引的
值标记为NaN（即pandas中的默认缺失值）。
'''
cs1=pd.Series([1.5,2.5,3,5,1],index=['a','c','d','b','e'])
cs2=pd.Series([10,20,30,50,10,100,20],index=['c','a','e','b','f','g','d'])
print(cs1,"\n",cs2)
```

```python
cs1+cs2
```

```python
import numpy as np
cdf1=pd.DataFrame(np.arange(10).reshape(2,5),columns=list('bcaed'))
cdf1
```

```python
import numpy as np
cdf1=pd.DataFrame(np.arange(10).reshape(2,5),columns=list('bcaed'),index=['no.1','no.2'])
cdf1
```

```python
cdf2=pd.DataFrame(np.arange(12).reshape(3,4),columns=list('abcd'))
cdf2
```

```python
cdf1+cdf2 #缺失值NaN与任何值相加的结果均为NaN
```

```python
'''
可以使用add、sub、div、mul等方法，将其他DataFrame对象的值传入指定DataFrame对象：

DataFrame.add(other, axis='columns', level=None, fill_value=None)
等价于dataframe + other，但是支持用fill_value替换其中一个输入中缺失的数据。

参数说明：
other : scalar, sequence, Series, 或 DataFrame
任何单个或多个元素数据结构，或类似列表对象。

axis : {0 或 ‘index’, 1 或 ‘columns’}
通过index(0 或 ‘index’)或columns(1 or ‘columns’)进行比较。
对于Series输入，axis匹配Series索引。

level : int 或 label
跨level广播，匹配传递的MultiIndex leve不匹配的索引将被合并在一起l的索引值。

fill_value : float 或 None, 默认 None
在计算之前，用这个值填充现有的缺失值(NaN)，以及成功进行DataFrame对齐所需的任何新元素。
如果两个对应的DataFrame位置中的数据都丢失，则结果将丢失。

不匹配的索引将被合并在一起。
'''
cdf1.add(cdf2,fill_value=0) 
```

#### 　　3.2.2.4 函数应用与映射

```python
'''
numpy的ufunc也可以应用于pandas对象，即可将函数应用到Series中的每一个元素。

apply、applymap和map方法均可对对象中的数据传递函数操作，区别如下：
-- apply的操作对象是DataFrame的一列或一行数据
-- applymap是元素级的，作用于每个DataFrame的每个数据
-- map也是元素级的，对Series中的每个数据调用一次函数
'''
reversef=lambda x:-x
reversef(cs2)
```

```python
rangef=lambda x:x.max()-x.min()
rangef(cs2)
```

```python
rangef(cdf1.add(cdf2,fill_value=0)) #对DataFrame使用函数默认是对列使用
```

```python
(cdf1.add(cdf2,fill_value=0)).apply(rangef,axis=0) #axis=0表示将函数应用于列
```

```python
(cdf1.add(cdf2,fill_value=0)).apply(rangef,axis=1) #axis=1表示将函数应用于行
```

```python
'''
传递给apply方法的函数还可以返回多个值组成的Series，也可以使用applymap方法格式化各个元素：
'''
def statistics(x):
    return pd.Series([x.min(),x.max(),x.max()-x.min(),x.mean(),x.count()],
                 index=['Min','Max','Range','Mean','N'])

(cdf1.add(cdf2,fill_value=0)).apply(statistics)
```

```python
outformat=lambda x: '%.2f' % x

((cdf1.add(cdf2,fill_value=0)).apply(statistics)).applymap(outformat)
```

```python
'''
pandas可以使用groupby方法对数据进行分组分析。
'''
jddf=pd.read_csv('data/data.csv',header=None,
                 names=['name','time','opening_price','closing_price',
                       'lowest_price','highest_price','volume'])
# header=None 表示不会自动把数据的第1行和第1列设置成行、列索引
# names指定列索引，即通常意义下的变量名
jddf.head()
```

```python
#课堂练习
# 生成“Market”变量：
#收盘价-开盘价，如果>0，取值为'Good'，<0，取值为'Bad'，=0，取值为‘Ok’。
```

```python
diff=jddf['closing_price']-jddf['opening_price']
diff
```

```python

m1=lambda x:'Good' if x>0 else ('Bad' if x<0 else 'OK')
m=list(map(m1,diff))
m
```

```python
jddf['Market']=m
jddf
```

```python

jddf['Market']=list(map(lambda x:'Good' if x>0 else ('Bad' if x<0 else 'OK'),jddf['closing_price']-jddf['opening_price']))
jddf
```

#### 　　3.2.2.6 合并

```python
'''
pandas提供了诸如concat、append、merge、join、combine_first、update、merge_ordered
merge_asof等函数或方法对pandas的数据对象进行合并。
'''
c1=pd.DataFrame({'Name':{101:'Zhang San',102:'Li Si',103:'Wang Laowu',
                        104:'Zhao Liu',105:'Qian Qi',106:'Sun Ba'},
                 'Subject':{101:'Literature',102:'History',103:'English',
                        104:'Maths',105:'Physics',106:'Chemics'},
                 'Score':{101:98,102:76,103:84,104:70,105:93,106:83}})
c1
```

```python
c2=pd.DataFrame({'Gender':{101:'Male',102:'Male',103:'Male',
                           104:'Female',105:'Female',106:'Female'}})
c2
```

```python
'''
concat主要是根据索引进行行或列的拼接，只能取行或列的交集或并集。

pd.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
          
objs：Series，DataFrame或Panel对象的序列或映射
axis :默认为0，为按行联合拼接；1 为按列联合拼接
join: 为另一个轴的拼接方式，包括 inner,outer，outer为并集，inner为交集
'''
#按列联合拼接，行取并集
c=pd.concat([c1,c2],axis=1)
c
```

```python
c=pd.concat([c1,c2],axis=0)#按行联合拼接，列取并集，列索引默认升序输出
c
```

```python
'''
使用append方法将指定行追加到现有的pandas对象中：
'''
c1.append(c2) #按行联合拼接，列取并集，列索引按升序排列
#该语句与pd.concat([c1,c2],axis=0)得到的结果相同
```

```python
'''
也可使用merge函数按照指定关键字进行合并：
merge可以将不同数据集依照某些字段（属性）进行合并操作，得到一个新的数据集。

pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)

left: 拼接的左侧DataFrame对象
right: 拼接的右侧DataFrame对象
how:默认为inner，基于列的数据内合并（取交集）,可设为inner/outer/left/right
on:要加入的列或索引级别名称。根据某个字段进行连接，必须存在于两个DateFrame中
   （若未同时存在，则需要分别使用left_on和right_on来设置）
'''
c3=pd.DataFrame({'Name':{101:'Zhang San',102:'Li Si',103:'Wang Laowu',
                        104:'Zhao Liu',105:'Qian Qi',106:'Sun Ba'},
                 'Gender':{101:'Male',102:'Male',103:'Male',
                           104:'Female',105:'Female',106:'Female'}})
c3
```

```python
#把c3与c1按照Name进行匹配合并，参加匹配合并的每一对象应具备同一个用来作为匹配标识的
#列（column):
pd.merge(c1,c3,on='Name')
```
