```python
from matplotlib import pyplot as plt

import numpy as np

plt.rcParams['font.sans-serif'] = ['kaiti']

plt.rcParams['axes.unicode_minus'] = False

plt.rcParams['figure.dpi'] = 150

plt.rcParams['figure.figsize'] = (5,4)

data = np.loadtxt("data.csv",dtype=np.float,skiprows=1,delimiter=',',usecols=(1))
print(data)
plt.hist(data, bins=10, edgecolor='white', color='blue')
plt.title("语文成绩")
plt.xlabel("分数段")
plt.ylabel("人数")
```
