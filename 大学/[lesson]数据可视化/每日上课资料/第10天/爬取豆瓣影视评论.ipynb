```python
import requests
from fake_useragent import UserAgent
ua = UserAgent(use_cache_server=False)


url = "https://movie.douban.com/subject/6424756/comments?limit=20&status=P&sort=new_score"

params = {
    "limit":20,
    "status":"P",
    "sort":"new_score",
}
headers = {
    "User-Agent":ua.random
}
res = requests.get(url,params=params,headers=headers)
print(res.text)
```

```python
from lxml import etree
# 将html转化为etree对象
html = etree.HTML(res.text)

comment_list = html.xpath('//p[@class=" comment-content"]/span/text()')
for comment in comment_list:
    with open("新蝙蝠侠.txt","a+",encoding="utf-8") as f:
        f.write(comment))
```

```python
## 总程序
import requests
import os
from fake_useragent import UserAgent
from lxml import etree
import time
ua = UserAgent(use_cache_server=False)

url = "https://movie.douban.com/subject/6424756/comments"

page_list = range(1,50)
# 判断文件是否存在,存在删除,路径使用的是相对路径
if os.path.exists("data/新蝙蝠侠评论.txt"):
    # 删除
    os.remove("data/新蝙蝠侠评论.txt")
# 循环1页数据
for page in page_list:
    params = {
        "limit":page * 20,
        "status":"P",
        "sort":"new_score",
    }
    headers = {
        "User-Agent":ua.random
    }
    res = requests.get(url,params=params,headers=headers)

    # 将html转化为etree对象
    html = etree.HTML(res.text)

    comment_list = html.xpath('//p[@class=" comment-content"]/span/text()')
    for comment in comment_list:
        with open("data/新蝙蝠侠评论.txt","a+",encoding="utf-8") as f:
            f.write(comment)
    # 限制采集时间,间隔3s
    time.sleep(3)
```
