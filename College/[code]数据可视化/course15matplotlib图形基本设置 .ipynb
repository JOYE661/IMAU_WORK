```python
#根据多项式绘制复杂图形： 
'''
-- 如果绘制更为复杂的函数图形，多项式是一种常用的解决方案。
-- 理论上可以使用多项式拟合任何曲线，但前提条件要事先对曲线的形式进行判断，
   即知道曲线的参数设置。
'''
```

<img src="./images/polynomial.png" width="50%">

```python
'''
numpy中通过设定变量x的各次幂（从左到右，从高到低的顺序）系数，利用poly1d函数
即可表示一个多项式函数：
'''
a=np.array([2,-3.5,1.6,-2,9])
p=np.poly1d(a)
print(p)
```

```python
#根据多项式p绘制一个多项式曲线：
#首先，创建x轴的数值，如在-10到10之间产生30个服从均匀分布的随机数：
x=np.linspace(-10,10,30)
print(x)
```

```python
#其次，利用多项式函数p创建多项式的值：
y=p(x)
print(y)
```

```python
#然后调用plot函数绘制图形，并用show函数显示：
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#在同一段matplotlib绘图语句中，可以将不同plot语句绘制的图形叠加在一个图形中：
#如绘制上述函数及其一阶导数叠加在一起的图形：
p1=p.deriv(m=1) #deriv方法的参数m为1得到一阶导函数
print(p1)
```

```python
y1=p1(x)
print(y1)
```

```python
plt.plot(x,y,
         x,y1)
plt.show()
```

```python
#设定所绘制线条和数据点的参数，包括颜色、线型、标记风格。
plt.plot(x,y,'b|',
         x,y1,'k-.')
#'b|'表示绘制以竖线|为数据点的蓝色曲线，’k-.'为黑色点线
plt.show()
```

```python
'''
可以使用xticks函数或yticks函数设置x,y轴的刻度。

plt.xticks(loc, labels): 
-- loc作用是给出刻度值
-- labels的作用是指定其显示形式，给出格式化转义方法 
   格式化转义需要在字符串首尾加r'$...$' （matplotlib中）
   pi需要转义才能显示为字符 pai. 若$-pi$ 则直接显示-pi
   分数的表示方式r'$-\frac{\pi}{2}$'
   如果没有第二个[]参数，刻度值显示如-3.142, -1.571...等浮点数，而不是-pi
'''
X=np.linspace(-np.pi,np.pi,256,endpoint=True)
C,S=np.cos(X),np.sin(X)
plt.plot(X,C,'--')
plt.plot(X,S) 
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi * 3 / 4,np.pi],
          [r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$',
           r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$',r'$\pi$'])
plt.yticks([-1, 0, +1],[r'$-1$',r'$0$',r'$+1$'])
plt.show()
```

```python
'''
matplotlib.pyplot.annotate()
功能：在图中带有指向型文本注释信息，用来显示在图形的一些特殊点,如极点, 最大值等。

参数信息：
s:str, 注释信息内容
xy:(float,float), 箭头点所在的坐标位置
xytext:(float,float), 注释内容的坐标位置
weight: str or int, 设置字体线型，
       其中字符串从小到大可选项有{‘ultralight’, ‘light’, ‘normal’, ‘regular’, 
       ‘book’, ‘medium’, ‘roman’, ‘semibold’, ‘demibold’, ‘demi’, ‘bold’,
       ‘heavy’, ‘extra bold’, ‘black’}
color: str or tuple, 设置字体颜色，tuple时用[0,1]之间的浮点型数据，RGB或者RGBA, 如:
      (0.1, 0.2, 0.5)、(0.1, 0.2, 0.5, 0.3)等
      
arrowprops：dict，设置指向箭头的参数，字典中key值有
①arrowstyle：设置箭头的样式，其value候选项如’->’,’|-|’,’-|>’,
              也可以用字符串’simple’,‘fancy’等。
②connectionstyle：设置箭头的形状，为直线或者曲线，候选项有’arc3’,‘arc’,‘angle’,
                ‘angle3’，可以防止箭头被曲线内容遮挡
③color：设置箭头颜色。

bbox：dict，为注释文本添加边框，其key有
①boxstyle,其格式类似’round,pad=0.5’，其可选项如下图所示
②facecolor(可简写为fc)设置背景颜色
③edgecolor(可简写为ec)设置边框线条颜色
④lineweight(可简写为lw)设置边框线型粗细
⑤alpha设置透明度，[0,1]之间的小数，0代表完全透明，即类似③颜色设置无效。
'''
```

```python
'''
这就是坐标矩阵——
横坐标矩阵X中的每个元素，与纵坐标矩阵Y中对应位置元素，共同构成一个点的完整坐标。
如B点坐标(X12,Y12)=(1,1)

X的每一行都一样，Y的每一列都一样。
基于这种强烈的规律性，numpy提供的numpy.meshgrid()函数可以让我们快速生成坐标矩阵X,Y。

语法：X,Y = numpy.meshgrid(x, y) 
输入的x，y，就是网格点的横纵坐标列向量（非矩阵）
输出的X，Y，就是坐标矩阵。
'''
```

```python
xnew = np.array([0, 1, 2])
ynew = np.array([1, 0])

X, Y = np.meshgrid(xnew, ynew)
print(X)
print(Y)

plt.plot(X, Y,
         color='red',  # 全部点设置为红色
         marker='.',  # 点的形状为圆点
         linestyle='')  # 线型为空，也即点与点之间不用线连接

plt.grid(True)
plt.show()
```

### 5.3.8 盒须图

```python
#是用一个类似盒子的图形来描述数据的分布状况
#matplotlib中的boxplot函数
```

<img src="./images/boxplot.png" width=60%>

```python
'''
盒图是在1977年由美国的统计学家约翰·图基(John Tukey)发明的。

它由五个数值点组成：最小值(min)，下四分位数(Q1)，中位数(median)，上四分位数(Q3)，最大值(max)。
也可以往盒图里面加入平均值(mean)。

下四分位数、中位数、上四分位数组成一个“带有隔间的盒子”。上四分位数到最大值之间建立一条延伸线，这个延伸线成为“胡须(whisker)”。

由于现实数据中总是存在各式各样地“脏数据”，也成为“离群点”，于是为了不因这些少数的离群数据导致整体特征的
偏移，将这些离群点单独汇出，而盒图中的胡须的两级修改成最小观测值与最大观测值

这里有个经验，就是最大(最小)观测值设置为与四分位数值间距离为1.5个IQR(中间四分位数极差)。即IQR = Q3-Q1，
即上四分位数与下四分位数之间的差，也就是盒子的长度。

最小观测值为min = Q1 - 1.5*IQR，如果存在离群点小于最小观测值，则胡须下限为最小观测值，离群点单独以点汇出。
如果没有比最小观测值小的数，则胡须下限为最小值。

最大观测值为max = Q3 -1.5*IQR，如果存在离群点大于最大观测值，则胡须上限为最大观测值，离群点单独以点汇出。
如果没有比最大观测值大的数，则胡须上限为最大值。

盒形图作用：
很好地指示数据中的值如何分布，尽管与直方图或密度图相比，盒形图似乎是原始的，但它们具有占用较少空间的优势，
这在比较许多组或数据集之间的分布时非常有用。
盒形图可能会有所不同，具体取决于数据的分布和样本的大小，例如，不对称且具有或多或少的异常值
 一批数据中的异常值是值得关注的，我们通过分析异常值产生的原因，往往能够发现解决问题的机会。
 可以根据中位数和平均值的相对位置大致判断数据的分布形态。
'''
```

### 认识jieba

```python
"""
因为中文文档与外文不同，字符之间没有空格，所以我们需要首先对中文的文本进行分词，通过分词获得单个的词语。

分词的原理：    
通过中文词库的方式来识别分词的
- 利用一个中文词库，确定汉字之间的关联概率
- 通过计算汉字之间的概率，汉字间概率大的组成词组，形成分词结果
- 除了分词，用户还可以添加自定义的词组

支持三种分词模式：
- 全模式，把句子中所有的可以成词的词语都扫描出来,存在冗余单词；
- 精确模式，试图将句子最精确地切开，适合文本分析，不存在冗余单词；
- 搜索引擎模式，在精确模式的基础上，对长词再次切分，适合用于搜索引擎分词。

支持繁体分词
支持自定义词典

中文的期刊、新闻等都是可以进行统计的。

"""
import jieba  #分词包
```

```python
#全模式
seg_list = jieba.cut("好好学习，天天向上",cut_all=True)
print( "/ ".join(seg_list)) #join()方法用于将序列中的元素以指定的字符连接生成一个新的字符串
```

```python
#精确模式
seg_list = jieba.cut("好好学习，天天向上",cut_all=False)
print ("/ ".join(seg_list)) 
```

```python
#搜索引擎模式
seg_list = jieba.cut_for_search("好好学习，天天向上") 
print ("/ ".join(seg_list))
```
